```python
import tsaug
print(dir(tsaug))
```

```python
import numpy as np
import matplotlib.pyplot as plt

filename = "output_data_ASI3.txt"  
data = np.loadtxt(filename, delimiter=" ")


time = data[:, 0]     
piezo = data[:, 1]
accel_z = data[:,2]
```

```python
plt.plot(time, piezo)
plt.ylim(-10,100)
plt.show()

plt.plot(time, accel_z)
plt.ylim(-10,100)
plt.show()
```

```python
from tsaug.visualization import plot
```

```python
plt.figure(figsize = (10,10))
plt.plot(time, piezo, color = 'r')
plt.ylim(0, 35)
#plt.xlim(0, 10)
plt.axhline(y = 1)
plt.show()
```

```python
## from tsaug import TimeWarp, Crop, Quantize, Drift, Reverse
my_augmenter = (tsaug.TimeWarp() * 1)

             #   + tsaug.Drift(max_drift=(0.4, 0.5)) @ 0.01 ) 
               
```

```python

piezo_aug= my_augmenter.augment(piezo)
plt.figure(figsize = (10,10))
plt.plot(time, piezo_aug)
plt.ylim(0, 35)
plt.axhline(y = 1)
#plt.xlim(0, 10)
plt.show()
```

```python
with open("Augmented.txt", "w") as f:
    for t, p in zip(time, piezo_aug):
        f.write(f"{t}, {p}\n")
        
```
