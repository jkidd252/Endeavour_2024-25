```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('RMS_data_wNoise_ASI3.txt', sep=" ", header=None)
data.columns = ["Time", "Vibrations", "Accel"]
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(data["Time"], data["Vibrations"])
ax2.plot(data["Time"], data["Accel"], color='red')
plt.show()
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(data["Time"], data["Accel"], color='blue')
ax1.scatter(data["Time"], data["Accel"], color='red')
ax2.plot(data["Time"], data["Vibrations"], color='red')

pre = data[ data["Time"] <= 1565 ]
ax1.scatter(pre["Time"], pre["Accel"], color='green')

ax1.set_ylabel('Accel [g]')
ax2.set_ylabel('Vibrations')
ax1.set_xlabel('Time [s]')
plt.title('RMS Dataset - Prelaunch')
plt.xlim(1560, 1570)
#plt.ylim(-5, 50)
plt.show()
```
